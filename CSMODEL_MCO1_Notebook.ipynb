## 1 Measures of Central Tendency

```python
# @title Create sample DataFrames

df1 = pd.DataFrame({"X": [54, 54, 54, 55, 56, 57, 57, 58, 58, 60, 60]})
df2 = pd.DataFrame({"X": [1, 2, 54, 54, 54, 55, 56, 57, 57, 58, 58, 60, 60, 200, 250]})
```

```python
# @title Mean, Median, Mode

mean1 = df1["X"].mean().round(1)
median1 = df1["X"].median()
mode1 = df1["X"].mode()[0]

print(f"[DF1] mean = {mean1}, median = {median1}, mode = {mode1}")

mean2 = df2["X"].mean().round(1)
median2 = df2["X"].median()
mode2 = df2["X"].mode()[0]

print(f"[DF2] mean = {mean2}, median = {median2}, mode = {mode2}")
```

```python
# @title Trimmed Mean

from scipy.stats import trim_mean

trimmed_mean2 = trim_mean(df2["X"], proportiontocut=0.15).round(1)

print(f"[DF2] trimmed mean (removed 15%) = {trimmed_mean2}")
```

## 2 Measures of Dispersion

```python
# @title Create sample DataFrames

df1 = pd.DataFrame({"X": [54, 54, 54, 55, 56, 57, 57, 58, 58, 60, 60]})
df2 = pd.DataFrame({"X": [1, 2, 54, 54, 54, 55, 56, 57, 57, 58, 58, 60, 60, 200, 250]})
```

```python
# @title Range

range1 = df1["X"].max() - df1["X"].min()
print(f"[DF1] range = {range1}")

range2 = df2["X"].max() - df2["X"].min()
print(f"[DF2] range = {range2}")
```

```python
# @title Interquartile Range

q1_1 = df1["X"].quantile(0.25)
q3_1 = df1["X"].quantile(0.75)
iqr_1 = q3_1 - q1_1
print(f"[DF1] IQR = {q3_1} - {q1_1} = {iqr_1}")

q1_2 = df2["X"].quantile(0.25)
q3_2 = df2["X"].quantile(0.75)
iqr_2 = q3_2 - q1_2
print(f"[DF2] IQR = {q3_2} - {q1_2} = {iqr_2}")
```

The IQR in the notebook differs from the ones in the slide due to how pandas computes for the quantile. See the `interpolation` parameter.

```python
# @title Variance and Standard Deviation

mean1 = df1["X"].mean()
variance1 = df1["X"].var()
std1 = df1["X"].std()
print(f"[DF1] mean = {mean1:.2f}, variance = {variance1:.2f}, standard deviation = {std1:.2f}")

mean2 = df2["X"].mean()
variance2 = df2["X"].var()
std2 = df2["X"].std()
print(f"[DF2] mean = {mean2:.2f}, variance = {variance2:.2f}, standard deviation = {std2:.2f}")
```

DF1 Interpretation
- The data points in DF1 are tightly clustered around the mean.
- The standard deviation of 2.25 means most values fall roughly within ±2.25 units of the mean (i.e., between about 54.4 and 58.9).
- Variance is low, which implies that DF1's data is more reliable/stable.

DF2 Interpretation
- DF2's data points are much more spread out around the mean.
- A standard deviation of 65.82 means values vary widely, typically within ±65.82 units of the mean (so roughly between 6 and 138).
- The very high variance confirms there is huge variability or inconsistency in DF2's data.
- DF2's data shows greater diversity or possibly the presence of outliers/extreme values

## 3 Correlation

```python
# @title Create sample DataFrames

data = {
    "height": [154, 160, 167, 152, 175, 161, 168],
    "weight": [60, 70, 65, 55, 85, 80, 75]
}

df = pd.DataFrame(data)
df
```

### 3.1 Pearson Correlation

$$r_{xy} = \frac{cov(x,y)}{\sigma_x \cdot  \sigma_y} = \frac{\sum_{i=1}^{n}(x_i-\overline{x})(y_i-\overline{y})}{\sqrt{\sum_{i=1}^{n}(x_i-\overline{x})^2} \cdot \sqrt{\sum_{i=1}^{n}(y_i-\overline{y})^2}}$$

```python
# @title Manual Computaion: Expand DataFrame

df["x_minus_mean"] = df["height"] - df["height"].mean()
df["y_minus_mean"] = df["weight"] - df["weight"].mean()
df["product"] = df["x_minus_mean"] * df["y_minus_mean"]
df["x_minus_mean_squared"] = df["x_minus_mean"] ** 2
df["y_minus_mean_squared"] = df["y_minus_mean"] ** 2
df
```

```python
# @title Manual Computation: Compute for Pearson Correlation

numerator = df["product"].sum()
denominator = np.sqrt(df["x_minus_mean_squared"].sum()) * np.sqrt(df["y_minus_mean_squared"].sum())
pearson = numerator / denominator

print(f"r = {pearson:.3f}")
```

```python
# @title Automatic Computation: Using pandas `corr()` function

pearson = df["height"].corr(df["weight"], method="pearson")
print(f"r = {pearson:.3f}")
```

### 3.2 Spearman Correlation

$$r_{xy} = \frac{cov(rank(x), rank(y))}{\sigma_{rank(x)} \cdot  \sigma_{rank(y)}} = 1 - \frac{6\sum{d_i^2}}{n(n^2-1)}$$

```python
# @title Manual Computaion: Expand DataFrame

df["rank_x"] = df["height"].rank()
df["rank_y"] = df["weight"].rank()
df["d"] = df["rank_x"] - df["rank_y"]
df["d_squared"] = df["d"] ** 2

df
```

```python
# @title Manual Computation: Compute for Spearman Correlation

numerator = 6 * df["d_squared"].sum()
denominator = len(df) * (len(df) ** 2 - 1)
spearman = 1 - (numerator / denominator)

print(f"r = {spearman:.3f}")
```

```python
# @title Automatic Computation: Using pandas `corr()` function

spearman = df["height"].corr(df["weight"], method="spearman")
print(f"r = {spearman:.3f}")
```
